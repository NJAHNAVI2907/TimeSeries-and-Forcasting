```python
# plot the autocorrelation of a differenced random walk

# create random walk
random_walk = df['S1']

# take difference
diff = list()
for i in range(1, len(random_walk)):
	value = random_walk[i] - random_walk[i - 1]
	diff.append(value)
# line plot
autocorrelation_plot(diff)
plt.show()
```
